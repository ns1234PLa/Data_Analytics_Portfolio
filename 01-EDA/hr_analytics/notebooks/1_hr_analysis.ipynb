```python
df_copy=df.copy()#safer copy
```

```python
#KPI's
#total headcount and attrition rate
total_emp= len(df_copy)
attrition_count=(df_copy['Attrition']== 'Yes').sum()
attrition_rate=(attrition_count/ total_emp)* 100

print("==================================================")
print(f"TOTAL HEADCOUNT: {total_emp} employees")
print(f"TOTAL ATTRITION: {attrition_count} left the company")
print(f"OVERALL ATTRITION RATE: {attrition_rate:.2f}%")
print("==================================================\n")
```

```python
#financial overview
avg_income = df_copy['MonthlyIncome'].mean()
median_income = df_copy['MonthlyIncome'].median()
print(f"Average Monthly Income: ${avg_income:,.2f}")
print(f"Median Monthly Income: ${median_income:,.2f}\n")
```

```python
#department breakdown
print("Department Performance Metrics:")
dept_analysis = df_copy.groupby('Department').agg(
    Total_Staff=('EmpID', 'count'),
    Avg_Monthly_Income=('MonthlyIncome', 'mean'),
    Departures=('Attrition', lambda x: (x == 'Yes').sum())
)
dept_analysis
```

```python
dept_analysis['Dept_Attrition_Rate_%'] = (dept_analysis['Departures'] / dept_analysis['Total_Staff']) * 100
print(dept_analysis.round(2))
print("\n")
```

```python
#experience and stability baseline
avg_tenure = df_copy['YearsAtCompany'].mean()
avg_total_experience = df_copy['TotalWorkingYears'].mean()
print(f"Average Loyalty: {avg_tenure:.1f} years at the company")
print(f"Average Industry Experience: {avg_total_experience:.1f} years\n")
```

```python
#data integrity check
missing_managers = df_copy['YearsWithCurrManager'].isnull().sum()
print(f"Data Quality Red Flag: Missing values in 'YearsWithCurrManager': {missing_managers}")
```

```python
#data cleaning
print(df_copy.isnull().sum())
```

```python
#relplace null with median
df_copy['YearsWithCurrManager'] = df_copy['YearsWithCurrManager'].fillna(df_copy['YearsWithCurrManager'].median())

print("\nMissing Values AFTER Cleaning:")
print(df_copy.isnull().sum())
```

```python
#remove duplicates
before_drop = len(df_copy)
df_copy.drop_duplicates(inplace=True)
after_drop = len(df_copy)
print(f"\nDuplicates removed: {before_drop - after_drop}")
```

```python
#relplace null with median
df_copy['YearsWithCurrManager'] = df_copy['YearsWithCurrManager'].fillna(df_copy['YearsWithCurrManager'].median())

print("\nMissing Values AFTER Cleaning:")
print(df_copy.isnull().sum())
```

```python
print("\nAttrition values:", df_copy['Attrition'].unique())
print("Department values:", df_copy['Department'].unique())
```

```python
df_copy.to_csv('../data/HR_Analytics_Cleaned.csv', index=False)
print("\n Cleaned data saved to: data/HR_Analytics_Cleaned.csv")
```
