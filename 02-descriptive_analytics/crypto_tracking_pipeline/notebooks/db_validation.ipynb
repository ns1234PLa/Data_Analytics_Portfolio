```python
import pandas as pd
import sqlite3
import os
```

```python
# connect to datawarehouse
base_dir = os.path.dirname(os.path.dirname(os.path.abspath('__file__')))
db_path = os.path.join(base_dir, "data", "crypto_warehouse.db")
conn = sqlite3.connect(db_path)
```

```python
#check tables
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Tables found in database:")
print(tables, "\n" + "-"*50)
```

```python
# shape and info for dim_assets
df_dim = pd.read_sql_query("SELECT * FROM dim_assets", conn)
print(f"Table: dim_assets")
print(f"   -> Rows: {df_dim.shape[0]} | Columns: {df_dim.shape[1]}")
print("\nColumn Data Types & Missing Values:")
print(df_dim.info())
print("\nSample Data:")
print(df_dim.head(2))
print("\n" + "="*50 + "\n")
```

```python
# shape and info for fact_market_status
df_fact = pd.read_sql_query("SELECT * FROM fact_market_status", conn)
print(f"Table: fact_market_status")
print(f"   -> Rows: {df_fact.shape[0]} | Columns: {df_fact.shape[1]}")
print("\nColumn Data Types & Missing Values:")
print(df_fact.info())
print("\nSample Data:")
print(df_fact.head(3))
```
